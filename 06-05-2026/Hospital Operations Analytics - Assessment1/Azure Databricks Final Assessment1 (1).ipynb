```python
daily_visits_data = [
(26,1002,201,"2024-03-14","Completed",2),
(27,1005,210,"2024-03-14","Completed",1),
(11,1011,205,"2024-03-06","Completed",3),
(20,1020,206,"2024-03-10","Completed",2),
(28,1018,203,"2024-03-15","Pending",2)
]
daily_visits_columns = ["visit_id","patient_id","doctor_id","visit_date","visit_status","tests_count"]
daily_visits_df = spark.createDataFrame(daily_visits_data, daily_visits_columns).withColumn("visit_date", to_date("visit_date"))
daily_visits_df.createOrReplaceTempView("daily_visits_updates")
```

```python
target = DeltaTable.forName(spark, "visits_delta_target")
target.alias("t").merge(
    daily_visits_df.alias("s"),
    "t.visit_id = s.visit_id"
).whenMatchedUpdate(set={
    "patient_id": "s.patient_id",
    "doctor_id": "s.doctor_id",
    "visit_date": "s.visit_date",
    "visit_status": "s.visit_status",
    "tests_count": "s.tests_count"
}).whenNotMatchedInsert(values={
    "visit_id": "s.visit_id",
    "patient_id": "s.patient_id",
    "doctor_id": "s.doctor_id",
    "visit_date": "s.visit_date",
    "visit_status": "s.visit_status",
    "tests_count": "s.tests_count"
}).execute()
```

**Why MERGE is used for incremental loads?**

MERGE is used for incremental loads because daily incoming data may contain both existing records and new records.
If a record already exists in the target table, MERGE updates the existing record with the latest values. If the record is new, MERGE inserts it into the table.
This helps avoid reloading the entire dataset every time new data arrives, making the process faster and more efficient.
MERGE is commonly used in ETL and data engineering pipelines where data is updated regularly, such as daily transactions, hospital visits, customer orders, or streaming data.

**Compare Parquet table and Delta table behavior**

Parquet is a columnar storage file format mainly used for storing and reading large amounts of data efficiently. It provides good compression and fast query performance, but it does not support advanced features like UPDATE, DELETE, MERGE, or Time Travel directly.

Delta Lake is built on top of Parquet and adds extra capabilities such as ACID transactions, transaction history, schema enforcement, Time Travel, and support for UPDATE, DELETE, and MERGE operations. Delta tables are more suitable for modern data engineering pipelines where data changes frequently.

**Why update is easier on Delta than raw Parquet**

Updating data is easier in Delta Lake because Delta supports direct row-level updates using SQL UPDATE statements.
In raw Parquet files, records cannot be updated directly. Usually the entire file or dataset must be read, modified, and rewritten again, which is more complex and time-consuming.
Delta Lake maintains transaction logs and metadata internally, allowing updates to happen safely and efficiently without manually rewriting all files.
