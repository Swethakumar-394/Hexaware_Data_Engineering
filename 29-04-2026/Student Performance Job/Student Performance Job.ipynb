```python
from pyspark.sql.functions import col, avg, when, count 
```

```python
data = [ 
(101, "Arjun Kumar", "Python", 86, 92), 
(102, "Neha Gupta", "Python", 74, 80), 
(103, "Rahul Nair", "Data Engineering", 88, 85), 
(104, "Priya Sharma", "AI", 91, 95), 
(105, "Siddharth Rao", "Data Engineering", 67, 72), 
(106, "Ayesha Khan", "AI", 79, 88), 
(107, "Kavya Reddy", "Python", 94, 96), 
(108, "Manish Yadav", "Data Engineering", 58, 69), 
(109, "Divya Menon", "AI", 83, 90), 
(110, "Farhan Ali", "Python", 62, 75) 
]
```

```python
columns = ["student_id", "student_name", "course", "marks", "attendance"] 
```

```python
df = spark.createDataFrame(data, columns) 
display(df)
```

```python
df = df.withColumn(
    "grade",
    when(col("marks") >= 90, "A")
    .when(col("marks") >= 75, "B")
    .when(col("marks") >= 60, "C")
    .otherwise("Needs Improvement")
)

display(df)
```

```python
df = df.withColumn(
    "certification_status",
    when(
        (col("marks") >= 75) & (col("attendance") >= 80),
        "Eligible"
    ).otherwise("Not Eligible")
)

display(df)
```

```python
avg_marks_per_course = df.groupBy("course").agg(
    avg("marks").alias("average_marks")
)

display(avg_marks_per_course)
```

```python
eligible_students_per_course = df.filter(
    col("certification_status") == "Eligible"
).groupBy("course").agg(
    count("*").alias("eligible_students_count")
)

display(eligible_students_per_course)
```
