Section 5 — Conditional Queries

```sql
SELECT *
FROM employees
WHERE salary > (SELECT AVG(salary) FROM employees);
```

```sql
SELECT e.*
FROM employees e
JOIN (
    SELECT department, AVG(experience) AS dept_avg_experience
    FROM employees
    GROUP BY department
) d
ON e.department = d.department
WHERE e.experience > d.dept_avg_experience;
```

Section 6 — Joins

```sql
SELECT e.emp_name, e.department, e.salary, p.project_name, p.budget
FROM employees e
JOIN projects p
ON e.department = p.department;
```

```sql
SELECT e.emp_name, e.department, p.project_name
FROM employees e
JOIN projects p
ON e.department = p.department;
```

```sql
SELECT department, SUM(budget) AS total_project_budget
FROM projects
GROUP BY department;
```

```sql
SELECT e.*
FROM employees e
WHERE e.department IN (
    SELECT DISTINCT department FROM projects
);
```

```sql
SELECT p.project_name, COUNT(e.emp_id) AS employee_count
FROM projects p
JOIN employees e
ON p.department = e.department
GROUP BY p.project_name;
```

Section 7 — Advanced Queries

```sql
SELECT *
FROM employees
WHERE department = 'Engineering'
ORDER BY salary DESC
LIMIT 3;
```

```sql
SELECT department, AVG(salary) AS average_salary
FROM employees
GROUP BY department
ORDER BY average_salary DESC
LIMIT 1;
```

```sql
SELECT city, COUNT(*) AS employee_count
FROM employees
GROUP BY city
ORDER BY employee_count DESC
LIMIT 1;
```

```sql
SELECT e.*
FROM employees e
JOIN (
    SELECT department, AVG(salary) AS dept_avg_salary
    FROM employees
    GROUP BY department
) d
ON e.department = d.department
WHERE e.salary > d.dept_avg_salary;
```

```sql
SELECT e.*
FROM employees e
JOIN (
    SELECT department, MAX(experience) AS max_experience
    FROM employees
    GROUP BY department
) d
ON e.department = d.department
AND e.experience = d.max_experience;
```

Final Mini Capstone

```sql
CREATE OR REPLACE TABLE employee_salary_analysis AS
SELECT
    e.emp_name,
    e.city,
    e.department,
    e.salary,
    e.experience,

    CASE
        WHEN e.salary >= 95000 THEN 'High'
        WHEN e.salary >= 80000 THEN 'Medium'
        ELSE 'Low'
    END AS salary_category,

    d.department_avg_salary

FROM employees e

JOIN (
    SELECT
        department,
        AVG(salary) AS department_avg_salary
    FROM employees
    GROUP BY department
) d
ON e.department = d.department;
```
