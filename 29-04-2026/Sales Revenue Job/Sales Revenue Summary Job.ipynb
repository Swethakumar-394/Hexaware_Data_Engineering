```python
from pyspark.sql.functions import col, sum
```

```python
data = [
    (1, "Aarav Sharma", "Hyderabad", "Laptop", 1, 75000),
    (2, "Priya Reddy", "Bengaluru", "Mouse", 4, 600),
    (3, "Rohan Mehta", "Mumbai", "Monitor", 2, 12000),
    (4, "Sneha Iyer", "Chennai", "Keyboard", 3, 1500),
    (5, "Kiran Patel", "Hyderabad", "Laptop", 1, 75000),
    (6, "Ananya Das", "Kolkata", "Tablet", 2, 30000),
    (7, "Vikram Singh", "Delhi", "Mouse", 6, 600),
    (8, "Meera Nair", "Bengaluru", "Monitor", 1, 12000),
    (9, "Rahul Verma", "Mumbai", "Laptop", 1, 75000),
    (10, "Pooja Shah", "Hyderabad", "Keyboard", 5, 1500)
]
```

```python
columns = ["sale_id", "customer_name", "city", "product", "quantity", "unit_price"]
```

```python
df = spark.createDataFrame(data, columns)
display(df)
```

```python
df = df.withColumn("revenue", col("quantity") * col("unit_price"))
display(df)
```

```python
result = df.groupBy("city").agg(
    sum("revenue").alias("total_revenue"))
display(result)
```

```python
result = df.groupBy("city").agg(
    sum("quantity").alias("total_quantity")
)
display(result)
```

```python
result = df.groupBy("city").agg(
    sum("revenue").alias("total_revenue"),
    sum("quantity").alias("total_quantity")
)
```

```python
result = result.orderBy(col("total_revenue").desc())
display(result)
```
