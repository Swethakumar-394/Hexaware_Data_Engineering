```python
#Final Combined Challenge
```

```python
#1

from pyspark.sql.functions import col, sum

# Load employees data
employees_df = spark.read.csv(" employees.csv", header=True, inferSchema=True)

# Load sales data
sales_df = spark.read.csv(" sales.csv", header=True, inferSchema=True)

# Add revenue column in sales
sales_with_revenue = sales_df.withColumn("revenue", col("quantity") * col("price"))

# Join employees and sales using emp_id
joined_df = employees_df.join(sales_with_revenue, on="emp_id", how="inner")

joined_df.show()
```

```python
#2

employee_revenue_df = joined_df.groupBy("name") \
    .sum("revenue") \
    .withColumnRenamed("sum(revenue)", "total_revenue") \
    .orderBy(col("total_revenue").desc())

employee_revenue_df.show()
```

```python
#3

employee_revenue_df.show(5)
```

```python
#4

department_revenue_df = joined_df.groupBy("department") \
    .sum("revenue") \
    .withColumnRenamed("sum(revenue)", "department_revenue") \
    .orderBy(col("department_revenue").desc())

department_revenue_df.show(1)
```

```python
#5

employee_revenue_df.coalesce(1).write.mode("overwrite").option("header", True).csv("final_sales_report.csv")
```
