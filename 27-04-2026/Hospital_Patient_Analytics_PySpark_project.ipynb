```python
patients.createOrReplaceTempView("patients")
doctors.createOrReplaceTempView("doctors")
appointments.createOrReplaceTempView("appointments")
bills.createOrReplaceTempView("bills")
```

```python
#1
spark.sql("""
SELECT *
FROM patients
""").show()
```

```python
#2
spark.sql("""
SELECT *
FROM patients
WHERE city = 'Hyderabad'
""").show()
```

```python
#3
spark.sql("""
SELECT city, COUNT(*) AS patient_count
FROM patients
GROUP BY city
""").show()
```

```python
#4
spark.sql("""
SELECT specialization, COUNT(*) AS doctor_count
FROM doctors
GROUP BY specialization
""").show()
```

```python
#5
spark.sql("""
SELECT *
FROM patients p
JOIN appointments a
ON p.patient_id = a.patient_id
""").show()
```

```python
#6
spark.sql("""
SELECT *
FROM doctors d
JOIN appointments a
ON d.doctor_id = a.doctor_id
""").show()
```

```python
#7
spark.sql("""
SELECT *
FROM appointments a
JOIN bills b
ON a.appointment_id = b.appointment_id
""").show()
```

```python
#8
spark.sql("""
SELECT payment_mode, SUM(bill_amount) AS total_bill_amount
FROM bills
GROUP BY payment_mode
""").show()
```

```python
#9
spark.sql("""
SELECT doctor_name, specialization, consultation_fee,
       RANK() OVER(ORDER BY consultation_fee DESC) AS rank
FROM doctors
""").show()
```

```python
#10
spark.sql("""
SELECT p.name, SUM(b.bill_amount) AS total_bill
FROM patients p
JOIN appointments a
ON p.patient_id = a.patient_id
JOIN bills b
ON a.appointment_id = b.appointment_id
GROUP BY p.name
ORDER BY total_bill DESC
LIMIT 1
""").show()
```
