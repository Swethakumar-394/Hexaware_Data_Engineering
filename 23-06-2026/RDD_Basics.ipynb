**RDD operations: two categories**

**1.Transformations**
These create a new RDD.
Examples:
map
filter
flatMap
distinct
union
intersection
groupByKey
reduceByKey

**2. Actions**
These return a result to the driver.
Examples:
collect
count
first
take
reduce


Map --> Transformation

Collected --> actions

**Map will allow you to itterate**

```python
rdd = sc.parallelize(["rahul"," sneha", "arjun"])

upper_rdd = rdd.map(lambda x: x.upper())

upper_rdd.collect()
```

**Filter** --> Transformation

It will keep the matching elements

Collection is the action


```python
rdd = sc.parallelize(["apple", "banana", "mango", "Kiwi"])

long_words = rdd.filter(lambda x: len(x) > 5)

long_words.collect()
```

**#FlatMap**

Map can catch only one element

**FlatMap** can catch multiple elements

```python
lines = sc.parallelize([
    "python spark hadoop",
    "spark sql pyspark"
])

words = lines.flatMap(lambda line: line.split(" "))

words.collect()
```

```python
lines = sc.parallelize(["hello world", "pyspark rdd"])

result = lines.map(lambda x: x.split(" "))

result.collect()
```

```python
lines = sc.parallelize(["hello world", "pyspark rdd"])

result = lines.flatMap(lambda x: x.split(" "))

result.collect()
```

```python
#Distinct Element

rdd = sc.parallelize([10, 20, 20, 30, 30, 30, 40])

rdd.distinct().collect()
```

```python
rdd1 = sc.parallelize([1, 2, 3])
rdd2 = sc.parallelize([4, 5, 6])

rdd1.union(rdd2).collect()
```

```python
rdd1 = sc.parallelize([1, 2, 3, 4])
rdd2 = sc.parallelize([2, 3, 4])

rdd1.intersection(rdd2).collect()
```

```python
rdd1 = sc.parallelize([1, 2, 3, 4])
rdd2 = sc.parallelize([3,4])

rdd1.subtract(rdd2).collect()
```

**MapValues**- only works on values

```python
rdd=sc.parallelize([
    ("Rahul",86),
    ("Sneha",76),
    ("Arjun",78)
])
rdd.mapValues(lambda x:x+2).collect()
```

```python
rdd=sc.parallelize([10,20,30,4])
rdd.sortBy(lambda x:x).collect()
```

```python
lines = sc.parallelize([
"python spark python",
"spark hadoop spark",
"python hadoop"
])
word_counts=(
    lines.flatMap(lambda x:x.split(" "))
    .map(lambda x:(x,1))
    .reduceByKey(lambda a,b:a+b)
)
word_counts.collect()
```

```python
employees = sc.parallelize([
("Rahul", 70000),
("Sneha", 60000),
("Arjun", 75000),
("Priya", 80000),
("Karan", 50000)
])
high_sal=employees.filter(lambda x:x[1]>65000)
high_sal.collect()
```
